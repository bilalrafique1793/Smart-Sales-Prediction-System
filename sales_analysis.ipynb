# Smart Sales Prediction System
## Data Analysis Module

This notebook loads `Advertising.csv`, cleans the dataset, checks data quality, saves the cleaned output, and generates all requested exploratory charts.

```python
from pathlib import Path
import sys
import pandas as pd
from IPython.display import display

project_root = Path.cwd()
if not (project_root / 'Advertising.csv').exists():
    project_root = project_root.parent

sys.path.insert(0, str(project_root / 'src'))

from data_loader import load_dataset, dataset_overview
from preprocessing import clean_dataset, dataset_quality_report, save_cleaned_dataset
from eda import create_eda_plots, generate_insights_markdown
```

## 1. Load Dataset

```python
raw_data = load_dataset(project_root / 'Advertising.csv')
overview = dataset_overview(raw_data)

print('Shape:', overview['shape'])
display(overview['head'])
print(overview['info'])
display(overview['describe'])
```

## 2. Clean Dataset

The cleaning step removes `Campaign_ID`, strips currency markers, removes extra spaces, standardizes `Region`, converts budgets and sales to floats, and drops duplicate rows.

```python
cleaned_data = clean_dataset(raw_data)
quality_report = dataset_quality_report(cleaned_data)
cleaned_path = save_cleaned_dataset(cleaned_data, project_root / 'data' / 'cleaned_advertising.csv')

display(cleaned_data.head())
display(quality_report['missing_values'])
print('Duplicate rows:', quality_report['duplicate_rows'])
print('Incorrect dtypes:', quality_report['incorrect_dtypes'])
print('Saved cleaned dataset to:', cleaned_path)
```

## 3. Exploratory Data Analysis

The following charts are saved to `outputs/graphs/`.

```python
graphs_dir = project_root / 'outputs' / 'graphs'
generated_graphs = create_eda_plots(cleaned_data, graphs_dir)
display(pd.DataFrame({'plot': list(generated_graphs.keys()), 'file': [str(path) for path in generated_graphs.values()]}))
```

The markdown below summarizes the main relationships observed in the data.
